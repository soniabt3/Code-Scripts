```python
##---------Type the code below this line------------------##
from sklearn.feature_selection import mutual_info_classif, chi2
import numpy as np
import pandas as pd

# Convert categorical columns to numerical using label encoding
X_encoded = np.abs(pd.get_dummies(X))
#X_encoded_abs = np.abs(X_encoded)

# Calculate mutual information
mutual_info = mutual_info_classif(X_encoded, y_encoded)

# Calculate chi-squared scores
chi2_scores, _ = chi2(X_encoded, y_encoded)

# Create a DataFrame to store the results
feature_scores = pd.DataFrame({'Feature': X_encoded.columns, 'Mutual Information': mutual_info, 'Chi-Squared Score': chi2_scores})

# Sort the DataFrame by mutual information score and chi-squared score in descending order
feature_scores.sort_values(by=['Mutual Information'], ascending=False, inplace=True)
feature_scores.sort_values(by=['Chi-Squared Score'], ascending=False, inplace=True)

# Display the top 5 significant features
print("Top 5 significant features based on Mutual Information:")
print(feature_scores.head(5)[['Feature', 'Mutual Information']])

print("\nTop 5 significant features based on Chi-Squared Score:")
print(feature_scores.head(5)[['Feature', 'Chi-Squared Score']])
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix, accuracy_score

# Assuming df3 is your DataFrame with the first column as the target variable and the remaining columns as features
class_y = df3.iloc[:, 0].values
class_X = df3.iloc[:, 1:].values

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(class_X, class_y, test_size=0.25, random_state=0)

# Scale the features
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)

# Initialize and train Logistic Regression classifier without specifying class weights
logistic_classifier = LogisticRegression(random_state=0, max_iter=1000)
logistic_classifier.fit(X_train, y_train)

# Predict values against the test set
logistic_y_pred_lr = logistic_classifier.predict(X_test)

# Compute and print accuracy for Logistic Regression
print("Logistic Regression")
logistic_cm_lr = confusion_matrix(y_test, logistic_y_pred_lr)
print(logistic_cm_lr)
logistic_accuracy = accuracy_score(y_test, logistic_y_pred_lr)
print("Accuracy:", logistic_accuracy)

print(" ")

# Initialize and train Logistic Regression classifier with balanced class weights to handle imbalanced dataset
weighted_logistic_classifier = LogisticRegression(random_state=0, max_iter=1000, class_weight='balanced')
weighted_logistic_classifier.fit(X_train, y_train)

# Predict values against the test set
weighted_logistic_y_pred = weighted_logistic_classifier.predict(X_test)

# Compute and print accuracy for Logistic Regression with balanced class weights
print("Logistic Regression with Weights to handle Imbalanced dataset")
weighted_logistic_cm = confusion_matrix(y_test, weighted_logistic_y_pred)
print(weighted_logistic_cm)
weighted_logistic_accuracy = accuracy_score(y_test, weighted_logistic_y_pred)
print("Accuracy:", weighted_logistic_accuracy)

print(" ")

# Initialize and train Logistic Regression classifier without specifying class weights
logistic_classifier = LogisticRegression(random_state=0, max_iter=1000, multi_class='multinomial', class_weight='balanced')
logistic_classifier.fit(X_train, y_train)

# Predict values against the test set
logistic_y_pred = logistic_classifier.predict(X_test)

# Compute and print accuracy for Logistic Regression
print("Multinomial Logistic Regression")
logistic_cm = confusion_matrix(y_test, logistic_y_pred)
print(logistic_cm)
logistic_accuracy = accuracy_score(y_test, logistic_y_pred)
print("Accuracy:", logistic_accuracy)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

# Assuming df3 is your DataFrame with the specified columns
column_names = ['price', 'latitude', 'longitude', 'bathrooms', 'bedrooms', 'square_feet', 'time']
column_positions = [df3.columns.get_loc(col) for col in column_names]
class_X = df3.iloc[:, column_positions].values

# Initialize and fit KMeans clustering
kmeans = KMeans(n_clusters=5)  # Example with 3 clusters
kmeans.fit(class_X)

# Print the inertia (within-cluster sum of squares)
print("Inertia (within-cluster sum of squares):", kmeans.inertia_)

# Calculate and print the silhouette score
silhouette_avg = silhouette_score(class_X, kmeans.labels_)
print("Silhouette Score:", silhouette_avg)

# Perform dimensionality reduction using PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(class_X)

# Plot the clusters in 2D
plt.figure(figsize=(8, 6))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=kmeans.labels_, cmap='viridis')
plt.title('Clusters Visualized using PCA')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.colorbar(label='Cluster')
plt.grid(True)
plt.show()
```

- Clustering can help identify distinct groups or segments within the rental market based on similar characteristics of apartments
- This can provide insights into different customer preferences, market trends, and potential business opportunities.
-  K-means is easy to understand and implement. It partitions the data into clusters based on similarity, making it intuitive for market segmentation.
-  K-means is computationally efficient and can handle large datasets with a large number of features, making it suitable for market segmentation tasks with potentially high-dimensional data.

- The performance can be measured by the Silhouette Score and Inertia.
- The Silhouette Score measures the quality of the clusters by interpreting it's closeness/likeness to it's cluster compared to other clusters.
- A score close to +1 indicates that the object is well matched to its own cluster and poorly matched to neighboring clusters.
- The K-means model returns a Silhouette Score of 0.9.
- PCA is used for dimentionality reduction so that the clusters can be represented in a two dimensional form.
